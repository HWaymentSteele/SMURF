```python
def parse_fasta(filename):
  '''function to parse fasta file'''    
  header, sequence = [],[]
  lines = open(filename, "r")
  for line in lines:
    line = line.rstrip()
    if len(line) == 0: pass
    else:
      if line[0] == ">":
        header.append(line[1:])
        sequence.append([])
      else:
        sequence[-1].append(line)
  lines.close()
  sequence = [''.join(seq) for seq in sequence]
  return header, sequence
```

```python
#Original function from Sergey. Only change is to output non one-hot sequences
def get_feat(filename, pad_to_max=False):
  '''
  Given A3M file (from hhblits)
  return MSA (aligned), MS (unaligned) and ALN (alignment)
  '''
  def parse_fasta(filename):
    '''function to parse fasta file'''    
    header, sequence = [],[]
    lines = open(filename, "r")
    for line in lines:
      line = line.rstrip()
      if len(line) == 0: pass
      else:
        if line[0] == ">":
          header.append(line[1:])
          sequence.append([])
        else:
          sequence[-1].append(line)
    lines.close()
    sequence = [''.join(seq) for seq in sequence]
    return header, sequence

  def one_hot(x,cat=None):
    if cat is None: cat = np.max(x)+1
    oh = np.concatenate((np.eye(cat),np.zeros([1,cat])))
    return oh[x]

  names, seqs = parse_fasta(filename)
  alphabet = "ACGU"
  a2n = {a:n for n,a in enumerate(alphabet)}
  def get_seqref(x):
    # input: string
    # output
    #   -seq: unaligned sequence (remove gaps, lower to uppercase, numeric(A->0, R->1...))
    #   -ref: reference describing how each sequence aligns to the first (reference sequence)
    n,seq,ref,aligned_seq = 0,[],[],[]
    for aa in list(x):
      if aa != "-":
        seq.append(a2n.get(aa.upper(),-1))
        if aa.islower(): ref.append(-1); n -= 1
        else: ref.append(n); aligned_seq.append(seq[-1])
      else: aligned_seq.append(-1)
      n += 1
    return np.array(seq), np.array(ref), np.array(aligned_seq)
  
  # get the multiple sequence alignment
  max_len = 0
  ms, aln, msa = [],[],[]
  for seq in seqs:
    seq_,ref_,aligned_seq_ = get_seqref(seq)
    if len(seq_) > max_len: max_len = len(seq_)
    ms.append(seq_)
    msa.append(aligned_seq_)
    aln.append(ref_)

  # pad each unaligned-sequence and alignment to same length
  if pad_to_max:
    for n in range(len(ms)):
      pad = max_len-len(ms[n])
      ms[n] = np.pad(ms[n],[0,pad],constant_values=-1)
      aln[n] = np.pad(aln[n],[0,pad],constant_values=-1)
  
  return msa, ms, aln
```

```python
def sub_sample(x, samples=1024, seed=0):
    np.random.seed(seed)
    idx = np.arange(1, len(x))
    idx = np.random.choice(idx, samples - 1, replace=False)
    idx = np.append(0, idx)
    return [x[i] for i in idx]
```

```python
def con_auc(true, pred, mask=None, thresh=0.01):
  '''compute agreement between predicted and measured contact map'''
  if mask is not None:
    idx = mask.sum(-1) > 0
    true = true[idx,:][:,idx]
    pred = pred[idx,:][:,idx]
  eval_idx = np.triu_indices_from(true, 2)
  pred_, true_ = pred[eval_idx], true[eval_idx] 
  L = (np.linspace(0.1,1.0,10)*len(true)).astype("int")
  sort_idx = np.argsort(pred_)[::-1]
  acc = [(true_[sort_idx[:l]] > thresh).mean() for l in L]
  return np.mean(acc)
```

## Set parameters

```python
num_fams = 1 #how many families from our data set do you want to use? 
mode = "EvolveMRF" # Standard, EvolveMRF, Gremlin
batch_sizes = [256,128,64] # bigger is usually better but memory constraints exist; feeding in a list will try sizes in that order
iters = "S" # S (2000 BasicAlign + 1000 MRF) or MB (extra 1000 iters on BasicAlign) or MM (extra 1000 iters on MRF); MM seems best
# Gremlin is best with batch_size 64 and lr = .05
```

```python
# options needed for SMURF
gradual = "no" # yes or no (yes = the consensus sequence is built from sequences "close to the query" and the definition of "close" evolves over time)
msa_frac = .3 # align to .3 * embedding of consensus seq + .7 * embedding of query seq
lr = .05
msa_memory =.90 # consensus = .95 consensus + .05 new alignment
```

```python
gap = -3
if mode == "EvolveMRF":
  MRF_steps = [1/4, 1/4, 1/4, 1/4]
  basic_steps = [1,0,0,0]
  if gradual == "yes":
    MRF_pid = [1.00, .60, .40, .20]
    init_align_to_msa_frac = 0.0
    init_msa_memory = False
  elif gradual == "no":
    MRF_pid = [0, 0, 0, 0]
    init_align_to_msa_frac = msa_frac
    init_msa_memory = msa_memory

if mode == "Gremlin":
    basic_steps = [0, 0, 0, 0]
    MRF_pid = [None, None, None, None]
    MRF_steps = [1, 0, 0, 0]
    
if mode == "Standard":
    basic_steps = [1, 0, 0, 0]
    MRF_pid = [None, None, None, None]
    MRF_steps = [1, 0, 0, 0]
    init_msa_memory = False
    init_align_to_msa_frac = 0.0

if iters == "XS":
    basic_steps = [int(200*_) for _ in basic_steps]
    MRF_steps = [int(100*_) for _ in MRF_steps]
if iters == "S":
    basic_steps = [int(2000*_) for _ in basic_steps]
    MRF_steps = [int(1000*_) for _ in MRF_steps]
if iters == "MB":
    basic_steps = [int(2000*_) for _ in basic_steps]
    MRF_steps = [int(2000*_) for _ in MRF_steps]
if iters == "MM":
    basic_steps = [int(3000*_) for _ in basic_steps]
    MRF_steps = [int(1000*_) for _ in MRF_steps]
```

## Run SMURF

```python
verbose = True
aucs = []
for fam in fams:
    if verbose:
        print(f"family {fam}")
    filename = 'a3m_files/' + fam + '.a3m'
    msa, ms, aln = get_feat(filename)
    a = {"ms": ms, "aln": aln}
    if len(a["ms"]) > 1024:
        seqs = nf.sub_sample(a["ms"])
        aln = nf.sub_sample(a["aln"])
    else:
        seqs = a["ms"]
        aln = a["aln"]
    ms = nf.one_hot(nf.pad_max(seqs))
    aln = nf.one_hot(nf.pad_max(aln))
    lens = np.array([len(seq) for seq in ms])

    for batch_size in batch_sizes:
        try:
            if verbose:
                print(f"attempting batch size {batch_size}")

            model = nf.BasicAlign(X=ms, lengths=lens, batch_size=batch_size, filters=512, win=18,
                                  sw_unroll=4, sw_temp=1.0, sw_learn_temp=False,
                                  sw_open=None, sw_gap=gap, sw_learn_gap=True,
                                  sw_restrict=False,
                                  seed=None, lr=lr, norm_mode="fast",
                                  w_scale=0.1)
            if verbose:
                print("BasicAlign")
            model.fit(basic_steps[0], verbose=verbose)
            msa_params = model.opt.get_params()

            if verbose:
                print("MRF")

            auc = []
            model = nf.MRF(X=ms, lengths=lens, ss_hide=0.15, batch_size=batch_size,
                           filters=512, win=18, lam=0.01,
                           sw_unroll=4, sw_temp=1.0, sw_learn_temp=False,
                           sw_open=None, sw_gap=None, sw_learn_gap=False,
                           nat_contacts=None,
                           nat_contacts_mask=None,
                           nat_aln=None, use_nat_aln=False, add_aln_loss=False, aln_lam=1.0,
                           seed=None, lr=lr, norm_mode="fast",
                           learn_bias=True, w_scale=0.1,
                           msa_memory=init_msa_memory, align_to_msa_frac=init_align_to_msa_frac, pid_thresh=MRF_pid[0])
            mrf_params = model.opt.get_params()
            for p in ["emb", "gap", "open"]:
                if p == "gap" and verbose:
                    print(p, msa_params[p])
                mrf_params[p] = msa_params[p]
            model.opt.set_params(mrf_params)
            model.fit(MRF_steps[0], verbose=verbose)

            if mode == "EvolveMRF":
                if verbose:
                    print(f"reset with pid thresh: {MRF_pid[1]}")
                model.reset_model_and_opt(
                    {"msa_memory": msa_memory, "align_to_msa_frac": msa_frac, "pid_thresh": MRF_pid[1]})
                model.fit(MRF_steps[1], verbose=verbose)

                if verbose:
                    print(f"reset with pid thresh: {MRF_pid[2]}")
                model.reset_model_and_opt(
                    {"msa_memory": msa_memory, "align_to_msa_frac": msa_frac, "pid_thresh": MRF_pid[2]})
                model.fit(MRF_steps[2], verbose=verbose)

                if verbose:
                    print(f"reset with pid thresh: {MRF_pid[3]}")
                model.reset_model_and_opt(
                    {"msa_memory": msa_memory, "align_to_msa_frac": msa_frac, "pid_thresh": MRF_pid[3]})
                model.fit(MRF_steps[3], verbose=verbose)

            print("X", mode, batch_size)

            with open('rna_out/contacts_' + fam, 'wb') as f:
                np.save(f, np.array(model.get_contacts()))

            break

        except:
            print(f"FAIL ON {fam} with batch size {batch_size}")
            auc = None
```

## Analyze EC

```python
# get lengths for each rfam

lens = {}
for fam in fams:
    filename = 'a3m_files/' + fam + '.a3m'
    seq = get_feat(filename)[0][0]
    lens[fam] = len(seq)
```

```python
# read in masks and true contacts

with open('rna_data/' + 'masks' + '.pkl', 'rb') as f:
    mas = pickle.load(f)

with open('rna_data/' + 'contact_pairs' + '.pkl', 'rb') as f:
    cons = pickle.load(f)
```

```python
# set minimum linear distance for contact map
min_d = 2
```

```python
#make masks

masks = {}
for fam in fams:
  masks[fam] = sparse.COO(mas[fam].T, 1, shape=(1000, 1000)).todense()

#RF02001, RF00504: trim pdb sequence with range given by Rfam
#pick submatrices for mask
masks['RF02001'] = masks['RF02001'][83:255, 83:255]
masks['RF00504'] = masks['RF00504'][82:168, 82:168]
for fam in fams:
  masks[fam] = masks[fam][:lens[fam], :lens[fam]]
```

```python
#Get true contact maps
contact_pairs = {}
for fam in fams:
  contact_pairs[fam] = sparse.COO(cons[fam], 1, shape=(1000, 1000)).todense()
#pick submatrices for contact map
contact_pairs['RF02001'] = contact_pairs['RF02001'][83:255, 83:255]
contact_pairs['RF00504'] = contact_pairs['RF00504'][82:168, 82:168]
for fam in fams:
  contact_pairs[fam] = contact_pairs[fam][:lens[fam], :lens[fam]]
```

```python
plt.rcParams.update({'font.size': 16})

def ppv(cutoff, con, con_true, min_d):
    """
    min_d -- pick contacts with minimum distance to diagonal
    """

    pos = np.argwhere(np.triu(con, k=1) >= cutoff)
    l = len(pos)
    n = 0
    for p in pos:
        if con_true.item(tuple(p)) == 1:
            n += 1
        else:
            pass

    return n / l

min_d = 6


def con_auc(true, pred, mask=None, thresh=0.01):
    '''compute agreement between predicted and measured contact map'''
    if mask is not None:
        idx = mask.sum(-1) > 0
        true = true[idx, :][:, idx]
        pred = pred[idx, :][:, idx]
    eval_idx = np.triu_indices_from(true, min_d)
    pred_, true_ = pred[eval_idx], true[eval_idx]
    L = (np.linspace(0.1, 1.0, 10) * len(true)).astype("int")
    sort_idx = np.argsort(pred_)[::-1]
    acc = [(true_[sort_idx[:l]] > thresh).mean() for l in L]
    return np.mean(acc)


class contacts_compare():

    def __init__(self, fam, contacts_s, contacts_g, contacts_true, mask, print_auc=False):
        self.print_auc = print_auc
        self.fam = fam
        self.mask = mask
        # get Gremlin contacts
        self.contacts_g = contacts_g
        self.L = len(self.contacts_g)
        self.contacts_g = ma.masked_array(self.contacts_g, self.mask)

        # get SMURF contacts
        self.contacts = contacts_s
        self.contacts = ma.masked_array(self.contacts, self.mask)

        # get true contacts
        self.contacts_t = contacts_true

        self.contacts_t = ma.masked_array(self.contacts_t, self.mask)
        self.resid1, self.resid2 = np.argwhere(contacts_true == 1).T

        self.ppv_E = []
        self.ppv_G = []
        for l in range(1, 2 * self.L):
            self.set_cutoff(l)
            self.ppvs = self.get_ppv()
            self.ppv_E.append(self.ppvs['SMURF'])
            self.ppv_G.append(self.ppvs['Gremlin'])
        self.set_cutoff(self.l)

    def auc(self):
        return {'SMURF': con_auc(pred=ma.getdata(self.contacts), true=ma.getdata(self.contacts_t), mask=self.mask),
                'Gremlin': con_auc(pred=ma.getdata(self.contacts_g), true=ma.getdata(self.contacts_t), mask=self.mask)}

    def set_cutoff(self, l):  # pick the top l contacts
        self.l = l
        self.cutoff = - \
            ma.sort(-self.contacts[np.triu_indices(self.L, k=min_d)])[self.l]
        self.cutoff_g = - \
            ma.sort(-self.contacts_g[np.triu_indices(self.L, k=min_d)])[self.l]

        # get contact pair indices, grouped in to true positives(tp) and false
        # positives(fp)
        self.con_coords_tp = np.argwhere(np.array((np.triu(self.contacts, k=min_d) > self.cutoff) * (
            self.contacts.mask == False) * (self.contacts_t == 1)) == True).T
        self.con_coords_fp = np.argwhere(np.array((np.triu(self.contacts, k=min_d) > self.cutoff) * (
            self.contacts.mask == False) * (self.contacts_t == 0)) == True).T
        self.con_coords_g_tp = np.argwhere(np.array((np.triu(
            self.contacts_g, k=min_d) > self.cutoff_g) * (self.contacts.mask == False) * (self.contacts_t == 1)) == True)
        self.con_coords_g_fp = np.argwhere(np.array((np.triu(
            self.contacts_g, k=min_d) > self.cutoff_g) * (self.contacts.mask == False) * (self.contacts_t == 0)) == True)
        self.con_coords_g_tp = self.con_coords_g_tp[:, ::-1].T
        self.con_coords_g_fp = self.con_coords_g_fp[:, ::-1].T

    def get_ppv(self):
        return {'SMURF': ppv(self.cutoff, self.contacts, self.contacts_t, min_d),
                'Gremlin': ppv(self.cutoff_g, self.contacts_g, self.contacts_t, min_d)}

    def plot2(self):
        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6.5))
        ax1.set_aspect(1)
        ax1.scatter(np.concatenate((self.resid1, self.resid2)),
                    np.concatenate((self.resid2, self.resid1)), color=".6")
        ax1.scatter(*self.con_coords_tp, facecolors='none', edgecolors='r')
        ax1.scatter(*self.con_coords_fp, marker="x", color='r')
        ax1.scatter(*self.con_coords_g_tp, facecolors='none', edgecolors='b')
        ax1.scatter(*self.con_coords_g_fp, marker="x", color='b')
        ax1.set_xlabel("Position 1")
        ax1.set_ylabel("Position 2")

        ax2.set_aspect(2 * self.L)
        ax2.plot(range(1, 2 * self.L), self.ppv_E, color='red', label='SMURF')
        ax2.plot(range(1, 2 * self.L), self.ppv_G,
                 color='blue', label='MLM-GREMLIN')
        ax2.set_ylim([0, 1.0 + .005])
        ax2.set_xlim([0, 2 * self.L])

        ax2.set_xlabel("Top N predictions")
        ax2.set_ylabel("PPV")
        ax2.legend(loc=(.1, .1))
        suptitle = fig.suptitle(self.fam, y=1.01)

        fig.savefig("figures/" + self.fam + ".pdf", dpi=fig.dpi,
                    bbox_inches='tight', bbox_extra_artists=[suptitle])
```

```python
auc_list=[]

for rfam in fams:

  with open('rna_out/contacts_g_'+ rfam, 'rb') as f:
    contacts_g = np.load(f)

  with open('rna_out/contacts_'+ rfam, 'rb') as f:
    contacts = np.load(f)[:lens[rfam], :lens[rfam]]

  mask = masks[rfam]
  contacts_t = contact_pairs[rfam]

  con = contacts_compare(rfam, contacts, contacts_g, contacts_t, mask)
  con.set_cutoff(con.L)
  con.plot2()
  auc_list.append(con.auc())
```

```python
#make scatter plot of auc values

auc_e = [fam['SMURF'] for fam in auc_list]
auc_g = [fam['Gremlin'] for fam in auc_list]

plt.rcParams.update({'font.size': 14, 'figure.autolayout': True})

fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(auc_g, auc_e, s=50, alpha=0.6,)
ax.set_xlim([.5,1])
ax.set_ylim([.5,1])
ax.set_xticks(np.arange(.5, 1.01, step=0.1))
ax.set_yticks(np.arange(.5, 1.01, step=0.1))

ax.set_xlabel('MLM-GREMLIN (AUC)')
ax.set_ylabel('SMURF (AUC)')
ax.plot([0,1], [0,1], 'k--', alpha=0.6, zorder=0)
ax.set_aspect(1)
ax.set_title('Protein Contact Accuracy')

ax.title.set_size(14)

plt.grid()

plt.savefig("figures/"+ "scatter" + ".pdf")
```

```python
#plot histogram of auc improvement

diff = np.array(auc_e) - np.array(auc_g)
med = np.median(diff).round(2)

plt.rcParams.update({'font.size': 14, 'figure.autolayout': True})

fig = plt.figure()
ax = fig.add_subplot(111)
ax.hist(diff, bins=11)
ax.set_xlim([-.25,.25])
ax.set_xticks(np.arange(-.2, .21, step=0.1))
ax.set_yticks(np.arange(0, 3.01, step=1))

ax.axvline(med, color='r', linestyle='dashed', linewidth=1)

ax.set_xlabel('SMURF improvement (AUC)')
ax.set_ylabel('Count')
ax.set_title('Median = ' + str(med))
ax.title.set_size(14)

ax.set_aspect(.15)

plt.savefig("figures/"+ "hist" + ".pdf")
```
